```python
import numpy as np
import matplotlib.pyplot as plt
import os 
import matplotlib.cm as cm

# Ottieni la directory corrente
current_directory = os.getcwd()

# Stampa la directory corrente
print("La directory corrente è:", current_directory)

# Cambia la directory corrente alla cartella 'Total_SISF'
os.chdir('Total_SISF')

# Ottieni la directory corrente
current_directory = os.getcwd()

# Stampa la directory corrente
print("La directory corrente è:", current_directory)
```

```python
# Mappa dei nomi dei file alle temperature corrispondenti
file_temperature_map = {
    "total_sisf_T_041.txt": 0.41,
    "total_sisf_T_043.txt": 0.43,
    "total_sisf_T_045.txt": 0.45,
    "total_sisf_T_047.txt": 0.47,
    "total_sisf_T_050.txt": 0.50,
    "total_sisf_T_052.txt": 0.52,
    "total_sisf_T_055.txt": 0.55,
    "total_sisf_T_060.txt": 0.60,
    "total_sisf_T_070.txt": 0.70,
    "total_sisf_T_080.txt": 0.80,
    "total_sisf_T_090.txt": 0.90,
    "total_sisf_T_1.txt": 1.0,
    "total_sisf_T_15.txt": 1.5,
    "total_sisf_T_2.txt": 2.0,
    "total_sisf_T_25.txt": 2.5,
    "total_sisf_T_30.txt": 3.0,
}

# Creiamo una figura per il plot
plt.figure(figsize=(10, 6))

# Normalizza le temperature per mappare i colori
norm = plt.Normalize(min(file_temperature_map.values()), max(file_temperature_map.values()))

# Iteriamo su ciascun file nella mappa
for file_name, temperature in file_temperature_map.items():
    # Leggi i dati dal file
    sisf = np.loadtxt(file_name)
    
    # Genera un array di tempo (supponendo che ogni punto corrisponda a un passo temporale uguale)
    time = np.arange(len(sisf)) * 0.05  # Supponendo un Delta t = 0.05
    
    # Ottieni il colore dalla colormap invertita
    color = cm.coolwarm_r(norm(temperature))
    
    # Plot SISF vs Time con colori e marker
    plt.plot(time, sisf, label=f'T = {temperature} K', color=color, marker='o', markersize=3)

# Aggiungi etichette e titolo al plot
plt.xlabel(r'$t$')
plt.ylabel('$F_s(q,t)$')
plt.title('Self-Intermediate Scattering Function for Different Temperatures')

# Aggiungi una linea orizzontale a F_s(q,t) = 1/e
plt.axhline(y=1/np.e, color='black', linestyle='--')

# Aggiungi la legenda
plt.legend()

# Mostra la griglia
plt.grid(True)

# Salva il plot come immagine PNG
plt.savefig('plot_SISF_Python_Simulation.png')

# Mostra il plot
plt.show()
```

```python
# Mappa dei nomi dei file alle temperature corrispondenti
file_temperature_map = {
    "total_sisf_T_041.txt": 0.41,
    "total_sisf_T_043.txt": 0.43,
    "total_sisf_T_045.txt": 0.45,
    "total_sisf_T_047.txt": 0.47,
    "total_sisf_T_050.txt": 0.50,
    "total_sisf_T_052.txt": 0.52,
    "total_sisf_T_055.txt": 0.55,
    "total_sisf_T_060.txt": 0.60,
    "total_sisf_T_070.txt": 0.70,
    "total_sisf_T_080.txt": 0.80,
    "total_sisf_T_090.txt": 0.90,
    "total_sisf_T_1.txt": 1.0,
    "total_sisf_T_15.txt": 1.5,
    "total_sisf_T_2.txt": 2.0,
    "total_sisf_T_25.txt": 2.5,
    "total_sisf_T_30.txt": 3.0,
}

# Creiamo una figura per il plot
plt.figure(figsize=(10, 6))

# Normalizza le temperature per mappare i colori
norm = plt.Normalize(min(file_temperature_map.values()), max(file_temperature_map.values()))

# Iteriamo su ciascun file nella mappa
for file_name, temperature in file_temperature_map.items():
    # Leggi i dati dal file
    sisf = np.loadtxt(file_name)
    
    # Genera un array di tempo (supponendo che ogni punto corrisponda a un passo temporale uguale)
    time = np.arange(len(sisf)) * 0.05  # Supponendo un Delta t = 0.05
    
    # Ottieni il colore dalla colormap invertita
    color = cm.coolwarm_r(norm(temperature))
    
    # Plot SISF vs Time con colori e marker
    plt.loglog(time, sisf, label=f'T = {temperature} K', color=color, marker='o', markersize=4)

# Aggiungi etichette e titolo al plot
plt.xlabel(r'$t$')
plt.ylabel('$F_s(q,t)$')
plt.title('Self-Intermediate Scattering Function for Different Temperatures')

plt.xscale('log')     # Logarithmic x-axis
plt.yscale('symlog')  # Symmetric logarithmic y-axis

# Aggiungi la legenda
plt.legend()

# Aggiungi una linea orizzontale a F_s(q,t) = 1/e
plt.axhline(y=1/np.e, color='black', linestyle='--')


# Mostra la griglia
plt.grid(True)

# Salva il plot come immagine PNG
plt.savefig('logplot_SISF_Python_Simulation.png')

# Mostra il plot
plt.show()
```
